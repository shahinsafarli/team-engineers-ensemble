# Interactive Experiment Explorer

Run `python src/experiments/run_all.py` first, then use the controls below to inspect generated Random Forest depth curves.

```python
from pathlib import Path
import csv
import matplotlib.pyplot as plt
from ipywidgets import interact, Dropdown

root = Path.cwd().parent if Path.cwd().name == 'notebooks' else Path.cwd()
rows = list(csv.DictReader((root / 'data' / 'results' / 'random_forest_depth.csv').open()))
datasets = sorted({row['dataset'] for row in rows})

def plot_depth(dataset):
    subset = [row for row in rows if row['dataset'] == dataset]
    depths = [int(row['max_depth']) for row in subset]
    test = [float(row['test_accuracy']) for row in subset]
    oob = [float(row['oob_accuracy']) for row in subset]
    plt.figure(figsize=(7, 4))
    plt.plot(depths, test, marker='o', label='Test accuracy')
    plt.plot(depths, oob, marker='s', label='OOB accuracy')
    plt.xlabel('Max depth')
    plt.ylabel('Accuracy')
    plt.title(dataset)
    plt.grid(alpha=0.25)
    plt.legend()
    plt.show()

interact(plot_depth, dataset=Dropdown(options=datasets, description='Dataset'));
```
